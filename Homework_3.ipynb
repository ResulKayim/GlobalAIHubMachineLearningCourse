```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
```

```python
# Generate dataset using make_classification function in the sklearn. 
# Convert it into pandas dataframe.
```

```python
X, y = make_classification(n_samples=1000, n_features=8, n_informative = 2, class_sep = 2, random_state = 42)

columns = [1, 2, 3, 4, 5, 6, 7, 8]
input_data = pd.DataFrame(data=X, index=range(1000), columns=columns)
```

```python
df = pd.DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
colors = {0:'red', 1:'blue', 2:'green', 3:'yellow', 4:'black'}
fig, ax = plt.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
```

```python
whole_data = input_data.copy()
whole_data['LABEL'] = y
categories = list(set(whole_data.LABEL))

sns.pairplot(whole_data)
```

```python
# Handle outliers (you can use IsolationForest, Z-score, IQR)
```

```python
from scipy import stats
import numpy as np
z = np.abs(stats.zscore(input_data))
z
```

```python
outliers = list(set(np.where(z > 3)[0]))
len(outliers)
```

```python
new_input = input_data.drop(outliers,axis = 0).reset_index(drop = False)
display(new_input)

y_new = y[list(new_input["index"])]
len(y_new)
new_input.drop(['index'], axis=1, inplace=True)
```

```python
whole_data['LABEL'].value_counts()
```

```python
# Split dataset into train and test set
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(new_input, y_new, test_size=0.3, random_state=42)
print(f"Number of train examples : {X_train.shape[0]} and number of test examples : {X_test.shape[0]}\n")
```

```python
# Import Decision Tree, define different hyperparamters and tune the algorithm.
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(max_depth=4, random_state=0)
clf.fit(X_train, y_train)
print("Accuracy of train:",clf.score(X_train,y_train))
print("Accuracy of test:",clf.score(X_test,y_test))
print("\nBias for training:",1-clf.score(X_train,y_train))
print("Bias for test:",1-clf.score(X_test,y_test))
print("\nVariance:",(1-clf.score(X_test,y_test))-(1-clf.score(X_train,y_train)))
```

```python
# Visualize feature importances.
```

```python
plt.figure(figsize=(23, 6))
importance = clf.feature_importances_
sns.barplot(x=importance, y=input_data.columns)
plt.show()
```

```python
# Create confusion matrix and calculate accuracy, recall, precision and f1 score.
```

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score, classification_report, f1_score
pred = clf.predict(X_test)
print(classification_report(y_test,pred))
```

```python
print("Precision = {}".format(precision_score(y_test, pred, average='macro')))
print("Recall = {}".format(recall_score(y_test, pred, average='macro')))
print("Accuracy = {}".format(accuracy_score(y_test, pred)))
print("F1 Score = {}".format(f1_score(y_test, pred,average='macro')))
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, pred)
plt.figure(figsize=(12, 8))
ax =sns.heatmap(cm, square=True, annot=True, cbar=False, fmt='g')
ax.xaxis.set_ticklabels(categories, fontsize = 12)
ax.yaxis.set_ticklabels(categories, fontsize = 12, rotation=0)
ax.set_xlabel('Predicted Labels',fontsize = 15)
ax.set_ylabel('True Labels',fontsize = 15)
```

```python
# Import XGBoostClassifier, define different hyperparamters and tune the algorithm.
```

```python
import xgboost as xgb

dmatrix_train = xgb.DMatrix(data=X_train, label=y_train)
dmatrix_test = xgb.DMatrix(data=X_test, label=y_test)

param = {'max_depth':4, 
         'eta':1, 
         'objective':'multi:softprob', 
         'eval_metric':'mlogloss',
         'num_class':2}

num_round = 5
model = xgb.train(param, dmatrix_train, num_round)
```

```python
#Predictions
preds = model.predict(dmatrix_test)
preds[:5]
```

```python
best_preds = np.asarray([np.argmax(line) for line in preds])
best_preds
```

```python
# Visualize feature importances.
```

```python
xgb.plot_importance(model)
```

```python
# Create confusion matrix and calculate accuracy, recall, precision and f1 score.
```

```python
from sklearn.metrics import confusion_matrix

plt.figure(figsize=(12, 8))
cm = confusion_matrix(y_test, best_preds)
ax = sns.heatmap(cm, square=True, annot=True, cbar=False, fmt='g')
ax.xaxis.set_ticklabels(categories, fontsize = 12)
ax.yaxis.set_ticklabels(categories, fontsize = 12, rotation=0)
ax.set_xlabel('Predicted Labels',fontsize = 15)
ax.set_ylabel('True Labels',fontsize = 15)
plt.show()
```

```python
print("Precision = {}".format(precision_score(y_test, best_preds, average='macro')))
print("Recall = {}".format(recall_score(y_test, best_preds, average='macro')))
print("Accuracy = {}".format(accuracy_score(y_test, best_preds)))
```

```python
from xgboost.sklearn import XGBClassifier
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV  

param_dict = {
    'max_depth':range(3,10,2),
    'min_child_weight':range(1,6,2),
    'learning_rate': [0.00001,0.001,0.01,0.1,1,2],
    'n_estimators': [10,190,200,210,500,1000,2000]
    
}

xgc = XGBClassifier(booster='gbtree', learning_rate =0.01, n_estimators=2000, max_depth=5,
 min_child_weight=1, gamma=0, subsample=0.8, colsample_bytree=0.8,
 objective= 'binary:logistic', nthread=4, eval_metric='mlogloss', seed=27)

clf = GridSearchCV(xgc,param_dict,cv=3, n_jobs = -1).fit(X_train,y_train)

print("\nTuned: {}".format(clf.best_params_)) 
print("Mean of the cv scores is {:.6f}".format(clf.best_score_))
print("Train Score {:.6f}".format(clf.score(X_train,y_train)))
print("Test Score {:.6f}".format(clf.score(X_test,y_test)))
print("Seconds used for refitting the best model on the train dataset: {:.6f}".format(clf.refit_time_))
```

```python
plt.figure(figsize=(12, 8))

xgb_pred = clf.predict(X_test)
cm = confusion_matrix(y_test, xgb_pred)
ax = sns.heatmap(cm, square=True, annot=True, cbar=False, fmt='g')
ax.xaxis.set_ticklabels(categories, fontsize = 12)
ax.yaxis.set_ticklabels(categories, fontsize = 12, rotation=0)
ax.set_xlabel('Predicted Labels',fontsize = 15)
ax.set_ylabel('True Labels',fontsize = 15)
```
